```python
target_film  = 'titanic'
subs_file = target_film + '/substitle.srt'
sl_video_path = target_film + '\SL'
```

```python
def create_prompts(subs_file):
    prompts = []
    with open(subs_file, 'r', encoding='iso8859_9') as f:
        chunks = f.read().split('\n\n')
        for chunk in chunks:
            if chunk == '':
                continue
            lines = chunk.split('\n')
            prompts.append(Prompt(lines[0], lines[1], lines[2:]))
    return prompts



def check_word(words,content):
    for word in words:
        if word in content.lower():
            return True
    return False


def write_data(l,name):
    file = open(name,'w')
    for s in l:
            file.write(s[0])
            file.write(' ')
            file.write(s[-1])
            file.write('\n')
    file.close()

def find_places(words, video_dif_list_delta, video_list):
    translator = str.maketrans('', '', string.punctuation)
    places = []
    index = 0
    for prompt in prompts:
        if prompt.content is not None:
            if check_word(words,prompt.content):
                start = convert_delta(prompt.interval[0].replace(',','.'))
                end = convert_delta(prompt.interval[1].replace(',','.'))
                duration = end - start
                vd,vd_start = select_video(start,video_dif_list_delta, video_list)
                c = prompt.content.strip()
                clean_sentence = ""
                if c != '':
                    clean_sentence  = c.translate(translator).lower()+" ."
                else:
                    continue
                sb = [vd,index,vd_start.total_seconds(),vd_start.total_seconds()+duration.total_seconds(),clean_sentence]
                #print(len(places),vd,vd_start,duration,clean_sentence)
                index += 1
                places.append(sb)
    print('NumberOfSentences:',len(places))
    random.shuffle(places)
    size = len(places)
    train_set = places[:int(size*0.7)]
    dev_set = places[int(size*0.7):int(size*0.85)]
    test_set = places[int(size*0.85):size]
    #write_data(train_set,'train.txt')
    #write_data(test_set,'test.txt')
    #write_data(dev_set,'dev.txt')
    return train_set,dev_set,test_set


def safe_create(path):
    try:
        shutil.rmtree(path)
        print(path,'Deleted')
        os.mkdir(path)
        print(path,'created.')
    except:
        os.mkdir(path)
        print(path,'created.')
```

```python
words = ['gel']
```

```python
from unidecode import unidecode
dicto = {}
p_anl = re.compile('\[.*\]')
with open('titanic.disamb.txt', 'r') as f:
    for idx, line in enumerate(f.readlines()):
        if line[:3] == '<S>' or line[:4] == '</S>':
            continue
        
        raw, anl = line.strip().split()[:2]
        #print(anl)
        m = p_anl.search(anl)
        #print(m)
        if m:
            if m.group() != '[Punc]':
                print(raw)
                dicto[unidecode(raw.lower())] = unidecode(anl[:m.span()[0]].lower())
        else:
            print('Warning in line {}: {}'.format(idx, line))
```

```python
new_resolution = 227
cut_sides = True


list_name = dev
list_type = 'dev'
dataset_path_tr = 'Dataset/' + list_type + '.tr'
dataset_path_sign = 'Dataset/' + list_type + '.sign'
file_sign = open(dataset_path_sign,'w')
file_tr = open(dataset_path_tr,'w')
cont = False
for i,vd in enumerate(list_name):
    dir_path =  target_film + '_' + str(vd[1])
    frame_path = 'Data/'+ dir_path
    subvideo_path = 'Clips/'+ dir_path+ ".mp4"
    word_list = vd[-1].split()
    word_list.pop(-1)
    print(word_list)
    for w in word_list:
        try:
            file_tr.write(dicto[unidecode(w)])
            file_tr.write(' ')
        except:
            print('error',w,unidecode(w))
            cont = True
            break
    if cont == True:
        cont = False
        continue
    safe_create(frame_path)
    file_tr.write('.\n')
    file_sign.write(frame_path)
    file_sign.write('\n')
    ffmpeg_extract_subclip(vd[0], vd[2]+1,vd[3]+1.8, targetname= subvideo_path)
    vidcap = cv2.VideoCapture('Clips/'+ dir_path+ ".mp4")
    success,image = vidcap.read()
    count = 0
    while success:
        if cut_sides == True:
            image = image[:,80:-80]
        image = cv2.resize(image, (new_resolution, new_resolution))
        d = 0.0001 * count
        cv2.imwrite(frame_path + '/frame' + str(d)[2:6] +'.png', image)     # save frame as JPEG file      
        success,image = vidcap.read()
        count += 1
file_tr.close()
file_sign.close()
```
